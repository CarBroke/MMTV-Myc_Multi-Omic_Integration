```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
# set a grey background (use sns.set_theme() if seaborn version 0.11.0 or above) 
sns.set_theme(style="whitegrid")

df = pd.read_csv("Myc_C6_PCA_Input.csv")

fig = plt.figure(figsize = (4.5,3.5))
fig = sns.violinplot(x='Histology', y='KRAS.LUNG_UP.V1_UP', data=df, saturation=1, order=["Microacinar", "Squamous", "EMT"], palette=["#EED21D","green","#050EC1"])
sns.swarmplot(x='Histology', y='KRAS.LUNG_UP.V1_UP', data=df, order=["Microacinar", "Squamous", "EMT"], color="black", size=5, alpha=0.4)
fig.set_xlabel("Tumor Subtypes", fontsize = 12, fontweight='bold')
fig.set_ylabel("KRAS.LUNG_UP.V1_UP", fontsize = 12, fontweight='bold')
fig = fig.get_figure()
sns.despine()
fig.savefig("KRAS_Violin_Plot.svg", dpi=1200, bbox_inches="tight")
sns.despine()
```
