```python
import numpy as np
import pandas as pd
from bioinfokit import analys, visuz
df = pd.read_csv("Top_Myc_Genes_ssGSEA_Test_Statistic_and_P-value.csv")
visuz.GeneExpression.volcano(df=df, lfc="Log2FC_EMT_Squam", pv="EMT_Squam_P-value",\
                             lfc_thr=(0.4, 0.4), pv_thr=(0.05,0.05), geneid="Name",\
                             genenames=("deg"),\
                             color=("red","grey","blue"), plotlegend=True, legendpos="upper right",\
                             legendanchor=(1.46,1), gstyle=1, gfont=12, sign_line=True, dim=(10,10),\
                             dotsize=15, legendlabels=["EMT Upregulated","No Difference","Squamous Upregulated"],\
                             show=True)
```
